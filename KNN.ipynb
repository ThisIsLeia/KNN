```python
# KNN分類法(K-Nearest Neighbor)
#電影分類
#尋找與玩命關頭相近的電影

import math
# 特徵值分別為[愛情親情, 跨國拍攝, 出現刀、槍, 飛車追逐, 動畫, 背景年代]
# 玩命關頭的特徵值
film = [5, 7, 8, 10, 2, 8]
film_titles = [
    '復仇者聯盟',
    '決戰中途島',
    '冰雪奇緣',
    '雙子殺手'
]

# 比較電影的特徵值
film_features = [
    [2, 8, 8, 5, 6, 10],
    [5, 6, 9, 2, 5, 6],
    [8, 2, 0, 0, 10, 2],
    [5, 8, 8, 8, 3, 8]
]

dist = []
for f in film_features:
    distance = 0
    for i in range(len(f)):
        distance += (f[i] - film[i]) ** 2
    dist.append(math.sqrt(distance))


min = min(dist)                     # 求最小值
min_index = dist.index(min)         # 最小值的索引

print("與玩命關頭最相似的電影 : ", film_titles[min_index])
print("相似度值 : ", dist[min_index])
for i in range(len(dist)):
    print("影片 : %s, 相似度 : %6.2f" % (film_titles[i], dist[i]))
```
